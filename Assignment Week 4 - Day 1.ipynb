## 1. How do you choose a machine learning algorithm for a problem statement? 
## Explain Briefly.

We have to choose suitable Machine Learning algorithm depending on the problem statement and the dataset and no model is better algorithm compared to another

Start with simple algorithum  logistic regression or SVM for classification problems. If these algorithms do fairly well, then the problem is simple, and pretty much every algorithm will work reasonably well.

If we do not get a satisfactory solution, then the problem might be harder, and you may have to use more advanced techniques. These would vary from problem to problem. For instance, a classification problem may be highly non-linear. So, you may need to do some non-trivial feature engineering for the simpler algorithms to work, use custom kernel functions for algorithms like SVM, or use deep learning for feature engineering (assuming you have sufficient data).

Another kind of hardness could be that the problem has too many features and too few training samples. In that case, you would mostly want to do some sort of feature selection, using techniques like sparse regression models (e.g. LASSO) or PCA.

Finally, a lot of this comes from experience. An expert would usually know in what order to try various algorithms for a specific kind of problem, such that you are roughly going from the simpler techniques to the more complicated ones. You would want to stop at a point where making the algorithm more complex doesn't give you much improvement in performance

## 2. What is Semi- Supervised Learning and why is it so important?


Semi-supervised machine learning is a combination of supervised and unsupervised learning. It uses a small amount of labeled data and a large amount of unlabeled data, which provides the benefits of both unsupervised and supervised learning while avoiding the challenges of finding a large amount of labeled data.

## 4. How do you treat heteroscedasticity in regression? What causes Heteroscedasticity and how to fix it?

In regression analysis, heteroscedasticity (sometimes spelled heteroskedasticity) refers to the unequal scatter of residuals or error terms. Specfically, it refers to the case where there is a systematic change in the spread of the residuals over the range of measured values.

Heteroscedasticity is a problem because ordinary least squares (OLS) regression assumes that the residuals come from a population that has homoscedasticity, which means constant variance.

When heteroscedasticity is present in a regression analysis, the results of the analysis become hard to trust. Specifically, heteroscedasticity increases the variance of the regression coefficient estimates, but the regression model doesn’t pick up on this.

There are three common ways to fix heteroscedasticity:

1. Transform the dependent variable
2. Redefine the dependent variable
3. Use weighted regression
